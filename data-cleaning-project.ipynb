```python
# Assuming both datasets have a 'City' column to merge on
merged_data = pd.merge(df, city_population_df, on='City')

merged_data.rename(columns = {'Yearly brutto salary (without bonus and stocks) in EUR': 'Salary'} , inplace = True)


merged_data.columns
```

```python

# Calculate average salary per capita
total_salary_by_city = merged_data.groupby('City')['Salary'].sum()
# to make this total salay a dataFram column we do these steps 
merged_data = merged_data.set_index('City')
merged_data['Total_Salary'] = total_salary_by_city

# Calculate the per capita salary
merged_data['Salary_Per_Capita'] = merged_data['Total_Salary'] / merged_data['2022 Population']

# Reset index if needed
merged_data = merged_data.reset_index()

# Display the result
new = merged_data[['City', 'Salary_Per_Capita']]
new[new['City'] != 'Berlin']
```
