Answer to the following Analytical questions: 

Question 1: Use Histograms to compare the disribution of movie `imdb_score` column related to movies produced in `France` and `Canada`. 

While creating the histgoram:
- Set the number of bins to 20 
- Set the xAxis Label 
- Set the yAxis Label
- Remove the grid lines


```python
# Create the histogram for Canada here 
movies_canada = data[data.country == 'Canada']

ax = movies_canada.imdb_score.hist(bins = 20)
ax.set_xlabel("Number of Movies")
ax.set_ylabel("Score")
ax.set_title("IMDB score of Movies made in Canada")
```

```python
# Create the histogram for France here 
movies_canada = data[data.country == 'France']

ax = movies_canada.imdb_score.hist(bins = 20)
ax.set_xlabel("Number of Movies")
ax.set_ylabel("Score")
ax.set_title("IMDB score of Movies made in France")
```
